```python
#importing important packages
import numpy as np
from math import sqrt
from math import exp
##########################################################################
k = 5           #k-value
N = 6     # number of time steps/steps in the path
m = 2*k+1       #no. of discrete space step
mass = 1.0   # mass of the particle
omega = 1.0  # angular frequency of the oscillator
hbar = 1.0   # Planck's constant divided by 2pi
q = []
p = []
xi = 0.0   # initial position
xf = 0.0    # final position
ti = 0.0    # initial time
tf = 1.0    # final time
epsilon=sqrt(2*np.pi/m)
dt = (tf-ti)/(N+1)
paths = []
curr_state = [xi, ]
 
for i in range(-k,k+1):
  q.append(i*epsilon)
  p.append(i*epsilon)
```

```python
#defining potential
def V(x):
    return 0.5 * mass * omega**2 * x**2 
```

```python
#creating all possible paths using dfs(depth first search) approach
def make_path(k, steps, paths, curr_path):
    if steps == 0:
        curr_path.append(xf)
        paths.append(curr_path.copy())
        curr_path.pop(-1)
        return
    
    for i in range(-k, k+1):
        curr_path.append(i*epsilon)
        make_path(k, steps-1, paths, curr_path)
        curr_path.pop(-1)
    return
make_path(k, N, paths, curr_state)
```

```python
#creaating the potential functional
c=complex(0,1)
def W(path):
  s = 0
  # for i in range(0,len(paths)):
  for j in range(0,N+2):
    s = s + V(path[j]) 
  s = np.exp(-c*s*dt)
  return s
```

```python
#creating the kernal function
def kernal(xi,xf):
  ker = 0
  for i in range(-k,k+1):
    ker=ker+(epsilon**2/(2*np.pi))*np.exp(c*((xi-xf)*i*epsilon-(i*epsilon)**2/2)*dt)
  return  ker
```

```python
#creating complex probability function

def cprob(path):
  p = 1
  for i in range(0,N+1):
    p = p*kernal(path[i],path[i+1])
  return p
```

```python
#creating complex amplitude function 
def ampl(paths):
  amp=0
  for i in range(0,len(paths)):
    amp = amp+cprob(paths[i])*W(paths[i])
  print(amp)
```
